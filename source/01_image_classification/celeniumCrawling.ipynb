```python
from selenium import webdriver
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
options.add_argument('--headless')        # Head-less 설정
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=options)
```

```python
url = "https://m.place.naver.com/restaurant/1857845585/review/visitor" 
# url = "https://m.place.naver.com/restaurant/1848424123/review/ugcc"
# url = "https://m.place.naver.com/restaurant/1848424123/review/visitor"
# url = "https://m.place.naver.com/restaurant/1121822928/review/visitor"
# url = "https://m.place.naver.com/restaurant/1514044516/review/visitor"
driver.get(url)
print(type(url))
DataDict = {}
```

```python
import time
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
time.sleep(2)

DataDict["별점"] = soup.select('div._20Ivz > span > em')[0].text.strip()
DataDict["방문자 리뷰 수"] = soup.select('div._20Ivz > span > a > em')[0].text.strip()
DataDict["블로그 리뷰 수"] = soup.select('div._20Ivz > span > a > em')[1].text.strip()
DataDict["리뷰ID"] = soup.select('div._1vou-')
DataDict["리뷰"] = soup.select('span.WoYOw')
print(DataDict)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
notices = soup.select('div._20Ivz > span > a > em')

for n in notices:
    print(n.text.strip())
```
