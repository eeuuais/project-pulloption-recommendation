### 내위치 설정 및 데이터 프레임 내 주변 1km 안쪽의 위치만 필터링

```python
newDf = pd.DataFrame(columns=['업소명', '위도', '경도', '거리'])
YongGu = (37.532804077918236,126.99089229965324 )
a = 0
my_map = folium.Map(location=YongGu, zoom_start=17)
for i in df.index :
  if(haversine(YongGu, (df["위도"][i], df["경도"][i]), unit = 'km') < 0.5) :
    newDf.loc[a] = [df["업소명"][i],df["위도"][i], df["경도"][i], haversine(YongGu, (df["위도"][i], df["경도"][i]), unit = 'km')]
    folium.Marker([df["위도"][i],  df["경도"][i]], tooltip=df["업소명"][i]).add_to(my_map)
    a = a + 1
    if a == 15 :
      break
folium.Marker(YongGu, tooltip="내 위치").add_to(my_map)
my_map.save("./gdrive/MyDrive/TeamPotato/지도데이터/mymap.html")

my_map.get_root().render()
my_map
```
